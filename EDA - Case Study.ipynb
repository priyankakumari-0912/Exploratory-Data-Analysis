### 2.3.1 Formatting the days fields -- DAYS_BIRTH, DAYS_EMPLOYED, DAYS_REGISTRATION

```python
df_curr_appl['Age'] = df_curr_appl['DAYS_BIRTH'].apply(lambda x: round(abs(x)/365))
df_curr_appl['YEARS_OF_WORK'] = df_curr_appl['DAYS_EMPLOYED'].apply(lambda x: round(abs(x)/365))
df_curr_appl['YEARS_OF_REGISTRATION'] = df_curr_appl['DAYS_REGISTRATION'].apply(lambda x: round(abs(x)/365))
```

#### Changing the value of YEARS_OF_WORK for Pensioneers to 0 as the days_employed is not a valid number for them

```python
df_curr_appl['YEARS_OF_WORK'] = df_curr_appl.YEARS_OF_WORK.apply(lambda x: 0 if x == 1001 else x)
```

#### Dropping the fields DAYS_BIRTH, DAYS_EMPLOYED, DAYS_REGISTRATION as they are no longer required

```python
df_curr_appl.drop(['DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION'], axis=1, inplace=True)
```
